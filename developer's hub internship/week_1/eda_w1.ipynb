## 05_manage null values using strategies

```python
Df=df.copy()
```

```python
# dropping columns with higher percentage of missing values
Df=Df.drop(columns=['deck'])
```

```python
Df=Df.drop(columns=['Unnamed: 0'])
```

```python
Df.columns
```

```python
# replacing missing values with mean , median or mode
Df['embark_town']=Df['embark_town'].fillna(Df['embark_town'].mode()[0])
```

```python
Df['embarked']=Df['embarked'].fillna(Df['embarked'].mode()[0])
```

```python
Df.isnull().sum()
```

```python
# checking uunique values in age column
Df['age'].unique()
```

```python
Df['age'].mean()
```

```python
Df['age']=Df['age'].fillna(Df['age'].mean())
```

```python
Df.isnull().sum()
```
